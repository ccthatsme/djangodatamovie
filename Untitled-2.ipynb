```python
import pandas as pd

dfLinks = pd.read_csv('moviecsv/links.csv')
dfTags = pd.read_csv('moviecsv/tags.csv')
dfMovies = pd.read_csv('moviecsv/movies.csv')
dfRatings = pd.read_csv('moviecsv/ratings.csv')


# dfLinks.head()
# dfTags.head()
dfMovies.head()
# dfRatings.head()
```

```python
movieWithRating = pd.merge(right=dfMovies, left=dfRatings, left_on='movieId', right_on='movieId', how='inner')

movieWithRating.head()
```

## Testing Boolean Mask By filtering all movies that released in 1991


```python


titleMask = movieWithRating['title'].str.contains('1991', case=False)

filterMask = movieWithRating[titleMask]

filterMask.head()
```

## Splitting title into movie title and year using a different method

```python
movieWithRating[['movie title','year']] = movieWithRating['title'].str.extract(r'(.*)\s\((\d{4})\)')

movieWithRating.drop('title', axis=1, inplace=True)

# dfMovies['Year'] = dfMovies['Year'].str[:-1].str[1:]


movieWithRating.dtypes
```

## Grouping by Year and Rating 

```python
movieGroupedWithYearAndStars = movieWithRating.groupby(
   ['year','rating']
).agg({'movie title': 'count'}).sort_values(by=['year', 'rating'], ascending=False)

movieGroupedWithYearAndStars.columns = ['Number_of_movies']

movieGroupedWithYearAndStars.reset_index()

movieGroupedWithYearAndStars.head(20)
```
