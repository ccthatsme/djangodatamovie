```python
import pandas as pd

dfLinks = pd.read_csv('moviecsv/links.csv')
dfTags = pd.read_csv('moviecsv/tags.csv')
dfMovies = pd.read_csv('moviecsv/movies.csv')
dfRatings = pd.read_csv('moviecsv/ratings.csv')


# dfLinks.head()
# dfTags.head()
dfMovies.head()
# dfRatings.head()
```

```python
movieWithRating = pd.merge(right=dfMovies, left=dfRatings, left_on='movieId', right_on='movieId', how='inner')

movieWithRating.head()
```

## Testing Boolean Mask By filtering all movies that released in 1991


```python


titleMask = movieWithRating['title'].str.contains('1991', case=False)

filterMask = movieWithRating[titleMask]

filterMask.head()
```

## Splitting title into movie title and year using a different method

```python
movieWithRating[['movie title','year']] = movieWithRating['title'].str.extract(r'(.*)\s\((\d{4})\)')

movieWithRating.drop('title', axis=1, inplace=True)

# dfMovies['Year'] = dfMovies['Year'].str[:-1].str[1:]


movieWithRating.head()
```

## Grouping by Year and Rating 

```python
movieGroupedWithYearAndStars = movieWithRating.groupby(
   ['year','rating']
).agg({'movie title': 'count'}).sort_values(by=['year', 'rating'], ascending=False)

movieGroupedWithYearAndStars.columns = ['Number_of_movies']

movieGroupedWithYearAndStars.reset_index()

movieGroupedWithYearAndStars.head(20)
```

```python
rating_counts = movieWithRating.groupby('year')['rating'].value_counts(normalize=True).mul(100)

movieGroupedWithYearAndStars = movieGroupedWithYearAndStars.join(rating_counts.round(2).astype(str).add('%'), on=['year','rating'])

movieGroupedWithYearAndStars = movieGroupedWithYearAndStars.rename(mapper={'proportion':'percentage_of_n_star_rated_movies_for_year'}, axis=1)

movieGroupedWithYearAndStars.reset_index()

movieGroupedWithYearAndStars.head(20)
```

## Year with the most 5 star Movies

```python
myMask = movieWithRating['rating'] == 5.0

movieWithHighestRating = movieWithRating[myMask].groupby(
   ['year','rating']
).agg({'movie title': 'count'}).reset_index()

movieWithHighestRating.columns = ['Year', 'Rating','Number_of_movies']

# movieWithHighestRating.reset_index()

movieWithHighestRating = movieWithHighestRating.sort_values(by=['Number_of_movies'], ascending=False)

movieWithHighestRating.head()

```

## Testing the Mean number of 5 star movies per year

```python
import torch
import torch.nn as nn
from torch.utils.data import TensorDataset, DataLoader
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np

numberOfMoviesWith5Stars = movieWithHighestRating['Number_of_movies']

mean5StarsPerYear = np.mean(numberOfMoviesWith5Stars)
print(mean5StarsPerYear)
```
